```python
import numpy as np
```

```python
rel = 'base_nom_0p8v/lib1_ff0p88v125c_base_400.tlib'
layer1 = '8v_case1'
```

```python
var = np.zeros( (3000, 14) )
for i in range(0, 3000):
    var[i] = np.load(layer1+'/a_x/'+str(i+1)+'.npy')
print(var)
```

```python
out = np.zeros( (3000, 49) )
for i in range(0, 3000):
    out[i] = np.load(layer1+'/a_y/'+str(i+1)+'.npy')
print(out)
```

```python
variable = np.transpose(var)
variable = variable[-3:]
variable = np.r_[[[1.0]*3000], variable]
scalar = np.linalg.inv(variable.dot(np.transpose(variable)))
scalar = scalar.dot(variable).dot(out)

print(scalar)
```

```python
testx = np.zeros( (186, 14) )
for i in range(0, 186):
    testx[i] = np.load(layer1+'/a_x/'+str(i+3001)+'.npy')
    
testx = np.transpose(testx)
testx = testx[-3:]
testx = np.r_[[[1.0]*186], testx]
testx = np.transpose(testx)

predict = testx.dot(scalar)
print(predict)
```

```python
ans = 0
for i in range(0, 186):
    testy = np.load(layer1+'/a_y/'+str(i+3001)+'.npy')
    predy = predict[i]
    for j in range(0, 49):
        real = float(testy[j])
        print(real)
        print(predy[j])
        ans = ans + min(1, abs(real-predy[j])/real)**2

import math
ans = 100-100*math.sqrt(ans/(186*49))
print(ans)
```
